```python
df = pd.DataFrame()

for i in range(1,500):
    response = requests.get(url=f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
    dataset = response.json()['results']
    df = pd.concat(objs=[df,pd.DataFrame(dataset)],ignore_index=True)
    df = df[['original_title','overview']]
    print('page - ',i)
```
