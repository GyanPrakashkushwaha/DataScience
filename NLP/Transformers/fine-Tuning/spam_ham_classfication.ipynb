```python
import pandas as pd
import os

# sklearn
from sklearn.preprocessing import LabelEncoder # to encode the data
from sklearn.model_selection import train_test_split

# Transformers
from transformers import DistilBertTokenizerFast

# tensorflow
import tensorflow as tf
from tensorflow.data import Dataset
```

```python
df = pd.read_csv(filepath_or_buffer=r'sentiment_analysis\sentiment_analysis_cleaned_dataset.csv')
```

```python
df = df.iloc[:1000]
```

```python
X = list(df['review'])
y = list(df['sentiment'])
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.50, random_state=42)
```

## Transformers

```python
X_train
```

```python
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
tokenizer
```

```python
train_encoding = tokenizer(X_train,truncation=True , padding=True)
```

```python
test_encoding = tokenizer(X_test,truncation=True , padding=True)
```

```python
train_encoding
```

```python
# Convert data encodings into dataset objects.
train_dataset = Dataset.from_tensor_slices((
            dict(train_encoding),
            y_train
        ))

test_dataset = Dataset.from_tensor_slices((
    dict(test_encoding),
    y_test
))
```

```python
import tensorflow as tf
from transformers import DistilBertTokenizerFast, TFDistilBertModel
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input

# Assuming you have X_train, y_train, X_test, y_test containing your data
# and the tokenizer is defined as you mentioned in the question.

# Create the DistilBERT model without the classification head (top layer).
base_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')

# Freeze the weights of the transformer layers so that they are not updated during training.
base_model.trainable = False

# Build the Sequential model.
model = Sequential()

# Add input layers to accept the tokenized inputs.
model.add(Input(shape=(None,), dtype=tf.int32, name='input_ids'))
model.add(Input(shape=(None,), dtype=tf.int32, name='attention_mask'))

# Pass the inputs through the DistilBERT model.
model.add(base_model)

# Add a dense layer for classification (adjust the number of units based on your task).
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model with an appropriate optimizer and loss function for your task.
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Convert data encodings into dataset objects.
train_dataset = tf.data.Dataset.from_tensor_slices((
    {'input_ids': train_encoding['input_ids'],
     'attention_mask': train_encoding['attention_mask']},
    y_train.reshape((-1, 1))  # Reshape y_train to (batch_size, 1)
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    {'input_ids': test_encoding['input_ids'],
     'attention_mask': test_encoding['attention_mask']},
    y_test.reshape((-1, 1))  # Reshape y_test to (batch_size, 1)
))

# Batch the datasets and enable prefetching for faster training.
batch_size = 32
train_dataset = train_dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.batch(batch_size)

# Train the model.
epochs = 5  # Adjust the number of epochs as needed.
model.fit(train_dataset, epochs=epochs)

# Evaluate the model on the test set.
eval_results = model.evaluate(test_dataset)
print("Test loss:", eval_results[0])
print("Test accuracy:", eval_results[1])
```

```python
from transformers import TFDistilBertForSequenceClassification , TFTrainer , TFTrainingArguments
```

```python
training_args = TFTrainingArguments(
    # output_dir='./results',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    # logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)
```

```python
with training_args.strategy.scope():
    model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = TFTrainer(
    model=model,                         # the instantiated ðŸ¤— Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)

trainer.train()
```
