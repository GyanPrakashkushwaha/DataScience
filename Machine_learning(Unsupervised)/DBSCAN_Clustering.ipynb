# DBSCAN Clustering

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_moons # for dataset
from sklearn.cluster import DBSCAN # this is my Algo
from sklearn.preprocessing import StandardScaler # to scale data
```

```python
X , _ = make_moons(n_samples=300,noise=0.05)
```

```python
X.shape
```

## Visualizing dataset

```python
plt.scatter(x=X[:,0],y=X[:,1],c=_)
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

## Visualizing dataset After scaling


```python
plt.scatter(
    x=X_scaled[:,0] , y=X_scaled[:,1] , c=_
)
```

## Applying Algo

```python
dbscan = DBSCAN(eps=0.5)
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

## Visualizing Clusters

```python
plt.scatter(x=X_scaled[:,0],y=X_scaled[:,1],c=dbscan.labels_)
```
