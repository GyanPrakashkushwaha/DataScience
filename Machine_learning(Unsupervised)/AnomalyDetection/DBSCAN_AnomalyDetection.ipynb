# DBSCAN Anomaly Detection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X ,_ = make_circles(n_samples=1000,noise=0.11,random_state=42,factor=0.3)
```

```python
X
```

### Visualizing data before Applying Algo 


```python
plt.scatter(x=X[:,0],y=X[:,1])
```

```python
dbscan = DBSCAN(eps=0.1,min_samples=6)
dbscan.fit(X)
```

```python
dbscan.fit_predict(X=X)
```

```python
dbscan.components_.shape
```

```python
print(dbscan.core_sample_indices_.shape)
print(dbscan.components_.shape)
```

```python
plt.scatter(x=X[:,0],y=X[:,1],c=dbscan.labels_)
```

```python
mask = dbscan.labels_ == -1
print(dbscan.labels_[mask])
print(dbscan.labels_[mask].shape)
```
