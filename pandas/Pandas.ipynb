### minimum and maximum values
we use min() and max for this

```python
# I wanna to know maximum fare
max(df['Fare'])
```

```python
# now I wanna to know that mans Details who has paid maximum fare
df['Fare'] == max(df['Fare'])
df[df['Fare'] == max(df['Fare'])]
```

```python
# now I wanna to know that mans only Name who has paid maximum fare
df[df['Fare'] == max(df['Fare'])]['Name']
```

```python
# I wanna to know minimum fare
min(df['Fare']) 
```

```python
# now I wanna to know that mans Details who has paid minimum fare
df['Fare'] == min(df['Fare'])
df[df['Fare'] == min(df['Fare'])]
```

```python
# now I wanna to know that mans only Name who has paid minimum fare
df[df['Fare'] == min(df['Fare'])]['Name']
```
