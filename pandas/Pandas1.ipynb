### ``merging`` Dataframes 
this work's in concept to sets

jiss bhii column ko merge krvana hota h vo dono Dataframes me hona chhiye nii to merge nii hoga on property se

```python
# making data to merge
df_merge1 = pd.DataFrame({
    'key1' : [1,2,3,4,5],
    'key2': [4,54,6,65,4],
    'key3': [4,5,2,3,5],
})
df_merge2 = pd.DataFrame({
    'key1' : [1,2,45,48,5],
    'key4' : [3,5,3,4,5],
    'key5': [2,36,2,33,453]
})
```

```python
df_merge1
```

```python
df_merge2
```

```python
# this is doing inner merge because it is default value of merging
pd.merge(df_merge1,df_merge2)
```

```python
pd.merge(df_merge1,df_merge2,how = 'left')
# its left is df_merge1 
# this is taking everything from left and matching with `key1` so whatever is not matching with key1 in right means in df_merge2 than retruning NaN
```

```python
pd.merge(df_merge1,df_merge2,how = 'right')
# this is taking df_merge2 for right
```

```python
pd.merge(df_merge1,df_merge2,how = 'outer')
# df_merge1 aur df_merge2 dono se data laa rha h aur jha bhii key1 ka data dono se match nii kr raa wha error de raa h..
```

```python
pd.merge(df_merge1,df_merge2,how='cross')
# ye  cartesian product create kr raa h
```

```python
# making data to use on property
df_merge3 = pd.DataFrame({
    'key2' : [1,2,3,4,5],
    'key9': [4,54,6,65,4],
    'key3': [4,5,2,3,5],
})
df_merge4 = pd.DataFrame({
    'key2' : [1,2,45,48,5],
    'key4' : [3,5,3,4,5],
    'key5': [2,36,2,33,453]
})
```

```python
pd.merge(df_merge3,df_merge4,on='key2')
# jiss bhii column ko merge krvana hota h vo dono Dataframes me hona chhiye nii to merge nii hoga on property se
# by default ye kye1 se match krke merge kr raa h but agr mujhe kissi aur key se match krvana h tb on = 'columnname ' de duga...
```

```python
df_merge1
```

### join function
ye join function merge ki trh hii km krta h but merge function pe koii column same hona chhiye aur iss join function function pe indexes same hone chhiye 
#### it means bhool gy 🤣🤣😂😂

syntax :  `df_join1.join(df_join2)`

```python
# making data for doing join operation
df_join1 = pd.DataFrame({
    'key2' : [1,2,3,4,5],
    'key9': [4,54,6,65,4],
    'key3': [4,5,2,3,5],
}, index= ['a','b','c','d','e'])

df_join2 = pd.DataFrame({
    'key22' : [1,2,45,48,5],
    'key4' : [3,5,3,4,5],
    'key5': [2,36,2,33,453]
},index= ['a','b','h','i','j'])
```

```python
df_join1
```

```python
df_join2
```

```python
df_join1.join(df_join2)
# this is doing left join by default
```

```python
df_join1.join(df_join2,how='right')
```

```python
df_join1.join(df_join2,how='outer')
```

```python
df_join1.join(df_join2,how= 'cross')
```
