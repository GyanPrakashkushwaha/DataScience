```python
plt.plot(x,y)
plt.xlabel('this is my x data')
plt.ylabel('this is my y lable')
plt.title('this is my graph')
```

```python
# to control plot's size
plt.figure(figsize=(7,4))
plt.plot(x,y,'--y')
plt.xlabel('this is my x data')
plt.ylabel('this is my y lable')
plt.title('this is my graph')
plt.grid()
```

```python
x1 = np.random.rand(50)
x1
y1 = np.random.rand(50)
```

```python
plt.scatter(x1,y1)
```

```python
# to give different colours to each data points
colours = np.random.rand(50)
# to give different sizes to each data point
sizes = 1000*np.random.rand(50)
plt.scatter(x=x1,y=y1,c=colours,s= sizes,alpha=.999)
plt.xlabel('this is my x lable')
plt.ylabel('this is my y lable')
plt.title('this is my scatter plot')
```

```python
# to give different colours to each data points
colours = np.random.rand(50)
# to give different sizes to each data point
sizes = 1000*np.random.rand(50)
plt.scatter(x=x1,y=y1,c=colours,s= sizes,alpha=.999,marker='*')
plt.xlabel('this is my x lable')
plt.ylabel('this is my y lable')
plt.title('this is my scatter plot')
```
