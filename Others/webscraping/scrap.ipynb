```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen 
import logging
```

```python
flipkart_URL = 'https://www.flipkart.com/search?q=' + 'iphone12pro'
```

```python
flipkart_URL
```

```python
url_client = urlopen(flipkart_URL)
```

```python
flipkart_page = url_client.read()
```

```python
flipkart_html = bs(flipkart_page,'html.parser')
```

```python
productlink = 'https://www.flipkart.com'+ bigbox[3].div.div.div.a['href']
```

```python
productlink
```

```python
product_req = requests.get(productlink)
product_req
```

```python
product_html = bs(product_req.text,'html.parser')
product_html
```

```python
all_reviews = product_html.findAll('div',{'class':'_16PBlm _3_IKGE'})
all_reviews
```
