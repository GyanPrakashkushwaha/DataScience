```python
backends = {
    'opencv': 'opencv',
    'ssd': 'ssd',
    'dblib': 'dblib',
    'mtcnn': 'mtcnn',
    'retinaface': 'retinaface',
    'mediapipe': 'mediapipe',
    'VGG-Face':'VGG-Face',
    'OpenFace':'OpenFace'
}
```

```python
for itmes,values in backends.items():
    try:
        face = DeepFace.detectFace(img_path=r'imgs\myimg_old.jpg',target_size=(224,224),detector_backend=f'{itmes}')
        plt.imshow(face)
        plt.title(values)
        plt.show()
    except:
        pass
    
```

```python
for itmes,values in backends.items():
    try:
        face = DeepFace.extract_faces(img_path=r'imgs\myimg_old2.jpg',target_size=(224,224),detector_backend=f'{itmes}')
        plt.imshow(face[0]['face'])
        plt.title(values)
        plt.show()
    except:
        pass
    
```
