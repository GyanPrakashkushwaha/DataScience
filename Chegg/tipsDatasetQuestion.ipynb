```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.scatterplot(data=df, x='tip', y='total_bill', hue='day', palette='bright')
plt.xlabel('Tips')
plt.ylabel('Total Bills')
plt.title('Scatter Plot of Tips vs Total Bills')
plt.legend(title='Day', loc='upper right')
plt.show()
```
