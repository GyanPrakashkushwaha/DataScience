###  5 Number Summary
      1. Minimum Value
      2. Q1- 25 percentile
      3. Median
      4. Q3- 75 percentile
      5. Maximum

```python
import numpy as np
import seaborn as sns
```

```python
## data
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100]
```

```python
# to calculate percentile
np.percentile(a=lst_marks,q=[0,25,50,75,100])
```

```python
# simple way to calculate 5 number summary
minimum,Q1,Q2,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
maximum
```

### calculating lower fence(boundary) and higher fence
if my data would be in range of these fence than my data would not be outlier othere wise my data points will be outliers


```python
# first need to calculate inter quartile range
IQR = Q3-Q1
IQR
```

```python
lower_fence = Q1-1.5*(IQR)
higher_fence = Q3+1.5*(IQR)
lower_fence,higher_fence
```
