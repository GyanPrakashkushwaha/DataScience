## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:
### there are few techniques
1 - Linear Interpolation

2 - Cubic Interpolation 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# for cubic interpolation
from scipy.interpolate import interp1d
```

## Linear Interpolation Code

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
sns.scatterplot(x=x,y=y)#both gives same output
plt.scatter(x=x,y=y) 
plt.show()
```

```python
## Interpolation data
# Above plot is showing linear relation between points And I wanna to make more data points between this data points
x_new = np.linspace(1,5,50) # creating new x values
# making  y interpolation values
y_interp = np.interp(x_new, # for interpolation I am giving this 
                     x,y # these are only for reference
                     )
```

```python
x_new, y_interp
```

```python
sns.scatterplot(x=x_new,y=y_interp)
```

## Cubic Interpolation 

```python
x = np.array([1,2,3,4,5])
y = np.array([  1,   8,  27,  64, 125])
```

```python
# this is creating cubic interpolation fucntion
func = interp1d(x=x,y=y,kind='cubic')
```

```python
func
```

```python
sns.scatterplot(x=x,y=y)
```

```python
## Interpolation data
x_new_cubic = np.linspace(1,5,50)
y_interp_cubic= func(x_new_cubic,)
```

```python
sns.scatterplot(x=x_new_cubic,y=y_interp_cubic)
```

## Polynomial Interpolation

```python
# creating sample data for polynomial interpolation
x = np.array([1,2,3,4,5])
y = np.array([  1,4,9,16,25])
```

```python
sns.scatterplot(x=x,y=y)
```

```python
# interpolating the data using polynomial interpolation
poly = np.polyfit(x,y,2) # fit a 2nd degree polynomial to the data
```

```python
x_new_poly = np.linspace(1,5,20) # new x values
y_interp_poly = np.polyval(poly,x_new_poly) # y Interpolated values
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new_poly,y_interp_poly)
```
