# UnitVectors
this converts all values in the range of  0 - 1

```python
import seaborn as sns
from sklearn.preprocessing import normalize # for Unit Vectors
import pandas as pd
import plotly.figure_factory as ff
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df.columns
```

```python
data = normalize(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
df_Vector = pd.DataFrame(data=data,columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

## Visualizing data after normalizing from Sklearn

```python
# before Scaling
fig = ff.create_distplot([df['sepal_length'],df['sepal_width'],df['petal_length'],df['petal_width']],
                         group_labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
fig
```

```python
# after scaling
fig = ff.create_distplot([df_Vector['sepal_length'],df_Vector['sepal_width'],df_Vector['petal_length'],df_Vector['petal_width']],
                         group_labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
fig
```
