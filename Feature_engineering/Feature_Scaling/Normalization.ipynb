## Normalization- Min Max Scaler
this is used in deep learning

### `tip`
if I will have some new values in my dataframe so simply I will have to transform my values in 2D array(means [[]] in this form). I will not have to fit that data

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler # for Normalization
import plotly.figure_factory as ff # for distplot
import pandas as pd
```

```python
df = sns.load_dataset('taxis')
```

```python
df.sample(10)
```

```python
df[['distance','fare','tip']]
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler
```

```python
min_max_scaler.fit(df[['distance','fare','tip']]) # Compute the minimum and maximum to be used for later scaling.
```

```python
data = min_max_scaler.transform(df[['distance','fare','tip']]) # this perform my MinMaxScaling by converting my values 0-1
# I can use min_max_scaler.fit_transform to directly do fit and transform
```

```python
dfs = pd.DataFrame(data=data,columns=['distance','fare','tip'])
dfs
```

## Visualizing data after Scaling from Sklearn

```python
# before Scaling
fig = ff.create_distplot([df['distance'],df['fare'],df['tip']],group_labels=['distance','fare','tip'])
fig
# here my data is between 0 - 140
```

```python
# after scaling
fig = ff.create_distplot([dfs['distance'],dfs['fare'],dfs['tip']],group_labels=['distance','fare','tip'],)
fig
# # here my data is between 0 - 1
```
