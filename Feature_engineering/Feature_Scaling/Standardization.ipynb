# Feature Scaling

## Standardization(Or Changing data to Standard Normal Distribution Or Z-Score)
this is used in Machine learning

```python
import seaborn as sns
import numpy as np
import plotly.figure_factory as ff # for distplot
from sklearn.preprocessing import StandardScaler # for Standardization
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

### I am gonna to Standardize my total_bill column

```python
total_bill_Standard = list(df['total_bill'])
len(total_bill_Standard)
```

```python
mean = np.mean(total_bill_Standard)
std = np.std(total_bill_Standard)
mean,std
```

```python
standardized_value=[]
for i in total_bill_Standard:
    z_score=(i - mean)/std
    standardized_value.append(z_score)
standardized_value
```

### visualizing both data

```python
# Before scaling
sns.distplot(df['total_bill'])
```

```python
# after scaling
sns.distplot(standardized_value)
```

```python
# before scaling
fig = ff.create_distplot([df['total_bill']],group_labels=['total_bill'])
fig
```

```python
# after scaling
fig = ff.create_distplot([standardized_value],group_labels=['standardized_value'])
fig
```

## now I will do Scaling with the help of sklearn lib

```python
scaler = StandardScaler()
scaler
```

```python
scaler.fit(df[['total_bill','tip','size']]) # It Compute the mean and std to be used for later scaling.
```

```python
Scaled_Data= scaler.transform(df[['total_bill','tip','size']]) # Perform standardization by centering and scaling. # basically this applies Z-Score formula
```

```python
Scaled_Data
```

```python
dfs =pd.DataFrame(Scaled_Data,columns=['total_bill','tip','size'])
```

## Visualizing data after Scaling from Sklearn

```python
# before scaling
x1 = df['total_bill']
x2 = df['tip']
x3 = df['size']

hist_data = [x1, x2,x3]

group_labels = ['total_bill', 'tip','size']


fig = ff.create_distplot(hist_data, group_labels)
fig.show()
```

```python
# after scaling
x1 = dfs['total_bill']
x2 = dfs['tip']
x3 = dfs['size']

hist_data = [x1, x2,x3]

group_labels = ['total_bill', 'tip','size']


fig = ff.create_distplot(hist_data, group_labels)
fig.show()
# looking nice😍
```
