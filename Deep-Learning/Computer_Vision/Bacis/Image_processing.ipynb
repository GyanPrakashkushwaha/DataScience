## Spltting The images channels
### means - dividing the Image into BGR.

```python
img = cv2.imread('data/Iron_man_hd.png')
B, G, R = cv2.split(m=img)

channels = {'B': B, 'G': G, 'R': R} 

for channel_name, channel_color in channels.items():
    if channel_name == 'B':
        cv2.imshow(f"Blue {channel_name}", channel_color)
    elif channel_name == 'G':
        cv2.imshow(f'Green {channel_name}', channel_color)
    else:
        cv2.imshow(f'Red {channel_name}', channel_color)

    cv2.waitKey()
    cv2.destroyAllWindows()

cv2.imshow(winname='original Image',mat=img)
cv2.waitKey()
cv2.destroyAllWindows()
```

## merge function

```python
img = cv2.imread('data/Iron_man_hd.png')
B, G, R = cv2.split(m=img)

merge = cv2.merge(mv=(R,G,B))
cv2.imshow(winname='merged- RGB Image',mat=merge)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(X=merge)
```
