# Exponential weighted moving average

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('DailyDelhiClimatetest.csv')
```

```python
df = df[['date', 'meantemp']]
```

```python
sns.scatterplot(data=df,x='date',y='meantemp')
```

### Alpha = 1 - beta
### beta = 1- alpha

## Beta = 0.1

```python
ewm = df['meantemp'].ewm(alpha=0.9).mean() # Here my beta is 0.1
df['ewm']=ewm
df
sns.scatterplot(data=df , y='meantemp',x='date')
sns.scatterplot(data=df , y='meantemp',x='date')
plt.plot(df['date'],ewm,color = 'black')
```

## Beta = 0.9

```python
ewm = df['meantemp'].ewm(alpha=0.1).mean() # Here my beta is 0.9
df['ewm']=ewm
sns.scatterplot(data=df , y='meantemp',x='date')
sns.scatterplot(data=df , y='meantemp',x='date')
plt.plot(df['date'],ewm,color = 'black')
```

## Beta = 0.5

```python
ewm = df['meantemp'].ewm(alpha=0.5).mean() # Here my beta is 0.5
df['ewm']=ewm
df
sns.scatterplot(data=df , y='meantemp',x='date')
sns.scatterplot(data=df , y='meantemp',x='date')
plt.plot(df['date'],ewm,color = 'black')
```
