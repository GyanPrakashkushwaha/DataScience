# Training Pre Trained Models - ResNet50

```python
from keras.applications.resnet import ResNet50 , preprocess_input , decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import cv2
from keras.utils import plot_model
import visualkeras
```

```python
model = ResNet50(weights='imagenet')
model.summary()
```

```python
plot_model(model=model)
```

```python
visualkeras.layered_view(model)
```

```python
def pred_visualize(img_path ='ImagesData/cat2.png'):
    img_visulize = cv2.imread(img_path)
    image_ =  plt.imshow(img_visulize)

    img = image.load_img(img_path,target_size=(228,228))
    x = image.img_to_array(img)
    x = np.expand_dims(x,axis=0)
    x = preprocess_input(x=x)
    predictions = decode_predictions(model.predict(x=x))
    return image_ , predictions
```

```python
import os
import random

def pred_visualize(img_path=None):
    img_dir='ImagesData'
    if img_path is None:
        img_files = [f for f in os.listdir(img_dir) if f.endswith('.png')]
        img_path = os.path.join(img_dir, random.choice(img_files))
    
    img_visulize = cv2.imread(img_path)
    image_ = plt.imshow(img_visulize)

    img = image.load_img(img_path, target_size=(228, 228))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x=x)
    predictions = decode_predictions(model.predict(x=x),top=3)
    
    return image_, predictions
```
